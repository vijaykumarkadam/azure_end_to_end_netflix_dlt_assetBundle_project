```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
dbutils.widgets.text("source_folder","")
dbutils.widgets.text("target_folder","")
```

```python
var_source = dbutils.widgets.get("source_folder")
var_target = dbutils.widgets.get("target_folder")
```

```python
df = spark.read.format("csv")\
        .option("header", "true")\
        .option("mode", "PERMISSIVE")\
        .load(f"abfss://bronze@storagefornetflixproject.dfs.core.windows.net/{var_source}")
```

```python
df.display()
```

```python
df.write.format("parquet")\
    .mode("append")\
    .option("path",f"abfss://silver@storagefornetflixproject.dfs.core.windows.net/{var_target}")\
    .saveAsTable(f"netflix_cata.silver.{var_target}")
```
