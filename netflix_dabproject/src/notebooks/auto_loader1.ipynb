```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
checkpoin_loc = 'abfss://bronze@storagefornetflixproject.dfs.core.windows.net/checkpoint'
```

```python
df = spark.readStream.format("cloudFiles")\
        .option("cloudFiles.format", "csv")\
        .option("cloudFiles.schemaLocation", checkpoin_loc)\
        .option("cloudFiles.schemaEvolutionMode", "rescue")\
        .option("header","true")\
        .load("abfss://rawdata@storagefornetflixproject.dfs.core.windows.net/")
```

```python
df.writeStream.format("csv")\
    .option("checkpointLocation", checkpoin_loc)\
    .trigger(once=True)\
    .option("header","true")\
    .start("abfss://bronze@storagefornetflixproject.dfs.core.windows.net/netflix_titles/")
```
